```python
#RESETTING THE INDEX AS WE CAN SEE WE HAVE TWO INDEX NUMBER FOR EXAMPLE(0 AND 6 FOR 1ST ROW)
df=df.reset_index()
df.head()
```

```python
#SAVING THE DATA IN NEW DF
df=df[['phonename','phoneprice','phonefeature']]
df.head()
```

```python
# VECTORIZATION OF DATA
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
#VECTORIZATION OF DATA AND CONVERTING IT TO ARRAY
vectors=cv.fit_transform( df['phonefeature']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
vectors[0]
```

```python
df.info()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(phone):
    phone_index=df[df['phonename']== phone].index[0]
    distances=similarity[phone_index]
    phones_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in phones_list:
        print(df.iloc[i[0]].phonename)
    
            
       
```

```python
df['phonename'].index[0]
```

```python
recommend('POCO C55 (Power Black, 64 GB)')




```
